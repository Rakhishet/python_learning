Numbers in Python

-> Python version 3 has 3 types of numbers int, float and complex numbers
-> Integers (int): The narrowest type.
-> Floating-point numbers (float): Wider than integers because any integer can be 
represented as a float (e.g., 5 becomes 5.0), but not all floats can be converted to integers without losing information
-> Complex numbers: The widest type. Any float can be represented as a complex number 
(e.g., 4.0 becomes 4.0 + 0j), but complex numbers cannot be converted to floats


```python
a = 67
print(type(a))
b = 3.56
print(type(b))
c = 2+2j
print(type(c))
```

The dir() function in Python is a built-in utility used to inspect the attributes and methods of objects.


```python
# When called without arguments, dir() lists all names in the current local scope:
print(dir())
# After importing modules, the list updates to include the new names:

import math
print(dir())
# Output: ['__annotations__', '__builtins__', '__doc__', '__name__', 'math', ...]

# With a Module
import random
print(dir(random))
# Output: ['Random', 'choice', 'randint', 'shuffle', 'uniform', ...]

# For built-in objects like lists or dictionaries, dir() lists their methods:

print(dir([]))
# Output: ['append', 'clear', 'copy', 'extend', 'pop', 'remove', 'sort', ...]
print(dir({}))
# Output: ['clear', 'copy', 'get', 'items', 'keys', 'update', 'values', ...]
```

 the help() function is a built-in utility that displays documentation about Python objects, modules, functions, classes, keywords, and more.

```python
# 1. Getting help on a built-in function
help(len)
# Getting help on a module
import math
help(math)
#3. Getting help on a class
help(dict)
# 4. Getting help on a keyword
help("for")
# 5. Starting the interactive help system
help()
```

```python
# Whats inside the datetime module
import datetime
dir(datetime)
# date class in datetime module
gvr = datetime.date(2001, 5, 22)
print(gvr)
print(gvr.strftime("%A, %B %d,%Y"))

launch_date = datetime.date(2017,2,27)
launch_time = datetime.time(22, 27)
launch_date_time = datetime.datetime(2017,2,27,22, 27)
print(launch_date_time)

# to get todays datetime
now = datetime.datetime.today()
print(now)

# converting string to date use datetime class strptime method
moon_landing = "7/20/1969"
moon_landing = datetime.datetime.strptime(moon_landing, "%m/%d/%Y")
print(moon_landing)
```

Python functions

Python Functions are a block of statements that does a specific task


```python
# Defining a Function
# We can define a function, using def keyword. 

def fun():
    print("Welcome to Coding")

# Calling a Function
# After creating a function, call it by using the name of the functions followed
# by parenthesis containing parameters of that particular function.

fun()
```

```python
# Function Arguments

# Types of Function Arguments

# 1. Default Arguments: A default argument is a parameter that assumes a default value if a value is 
# not provided in the function call for that argument.
# when we have positional argument and default agrument twogether default argument comes at the end

def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10)

# 2. Keyword Arguments: values are passed by explicitly specifying the parameter names
# , so the order doesn’t matter.

def student(fname, lname):
    print(fname, lname)

student(fname='Rakhi', lname='Shet')
student(lname='Shet', fname='Rakhi')

# 3. Positional Arguments: values are assigned to parameters based on their order in the function call.

def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Olivia", 27)

print("\nCase-2:")
nameAge(27, "Olivia")


# 4. Arbitrary Arguments: allow a function to accept a variable number of inputs. This is done using two special symbols:

# *args: collects extra positional (non-keyword) arguments as a tuple.
# **kwargs: collects extra keyword arguments as a dictionary.

def func(*positional_args):
    for i in positional_args:
        print(i)

func(10,20,40)

def func(**keyword_args):
    for k,v in keyword_args.items():
        print(k,v)
func(a=10,b ="happpy")
```

Logging in Python

Logging in Python is a powerful and flexible system for tracking events that occur while software runs. It allows developers to record messages that can help diagnose problems, understand program flow, and monitor the behavior of applications. 
The logging module in Python provides a way to configure different log handlers and formatters to control how log messages are outputted.

Purpose of Logging:
1. Debugging: Logging helps developers identify and fix issues in their code by providing detailed information about the program's execution.
2. Monitoring: Logs can be used to monitor the health and performance of applications in production environments
3. Auditing: Logs can serve as an audit trail, recording important events and actions taken by users or the system.

Levels of Logging:

1. DEBUG: Detailed information, typically of interest only when diagnosing problems.
2. INFO: Confirmation that things are working as expected.
3. WARNING: An indication that something unexpected happened, or some problem might occur in the near future.
4. ERROR: A more serious problem that prevents some function from working correctly.
5. CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.

```python
import logging

# Create and configure logger
# logger will set the message level to DEBUG(10), which means it will capture all messages at the INFO level and above (WARNING, ERROR, CRITICAL).
# default file mode is 'a' which means append, if we want to overwrite the file we can set filemode to 'w'
logging.basicConfig(filename="newfile.log"
                    , level = logging.DEBUG
                    , format = "%(asctime)s - %(levelname)s - %(message)s"
                    , filemode = 'w')
logger = logging.getLogger()

logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")

```
